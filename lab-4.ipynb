```python

def show_factorial(n):
    fact = 1
    steps = ""

    for i in range(n, 0, -1):
        fact *= i
        steps += str(i)
        if i != 1:
            steps += " × "

    print(f"{n}! = {steps} = {fact}")


show_factorial(5)
show_factorial(4)
show_factorial(3)
```

```python
# Create Car class
class Car:
    def __init__(self, brand):
        self.brand = brand

    def display(self):
        print("Car Brand:", self.brand)


# Create objects
c1 = Car("Toyota")
c2 = Car("BMW")
c3 = Car("Tesla")

# List of objects
car_list = [c1, c2, c3]

# Dictionary of objects
car_dict = {
    "car1": c1,
    "car2": c2,
    "car3": c3
}

# Loop through list
print("Display using List:")
for car in car_list:
    car.display()

# Loop through dictionary
print("\nDisplay using Dictionary:")
for key, value in car_dict.items():
    print("Key:", key)
    value.display()
```

```python
# Create Student class
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)
        print("-------------------")


# Create list of student objects
students = []
students.append(Student("Tanisha", 18, 85))
students.append(Student("Rahul", 19, 90))
students.append(Student("Aman", 18, 78))

# Loop to display all student objects
for student in students:
    student.display()
```

```python
# Create Student class
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)
        print("-------------------")


# List of student objects
students_list = []

# Dictionary of student objects (key = name)
students_dict = {}

# Creating objects
s1 = Student("Tanisha", 18, 85)
s2 = Student("Rahul", 19, 90)
s3 = Student("Aman", 18, 78)

# Adding to list
students_list.append(s1)
students_list.append(s2)
students_list.append(s3)

# Adding to dictionary
students_dict[s1.name] = s1
students_dict[s2.name] = s2
students_dict[s3.name] = s3

# Loop through list
print("Display using List:")
for s in students_list:
    s.display()

# Loop through dictionary
print("\nDisplay using Dictionary:")
for key, value in students_dict.items():
    print("Key:", key)
    value.display()
```

```python
# Create Employee class
class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Salary:", self.salary)
        print("-------------------")


# Create list of employee objects
employees = []
employees.append(Employee("Tanisha", 25, 30000))
employees.append(Employee("Rahul", 28, 40000))
employees.append(Employee("Aman", 26, 35000))

# Loop to display all employee objects
for emp in employees:
    emp.display()
```
